```python
#Question - 3

# to read input from the command line.
import sys
# spark configuration helps in naming spark, the how many cores to us and memory
#sparkcontextto create RDDs
from pyspark import SparkConf, SparkContext
# to be able to perform cosine similarity statistic
from math import sqrt

#load the movie data and store in dictionary
#defining the function
def loadMovieNames():
#creating dictionary
    movieNames = {}
#locating file in cloudera
    with open("/home/cloudera/moviedata/itemfile.txt") as f:
#scan through everyline
        for line in f:
#split by pipe character and stores the result in fields.
            fields = line.split('|')
#Converts the first field to an integer and uses it as the key in the dictionary, with the second field as the value
            movieNames[int(fields[0])] = fields[1].decode('ascii', 'ignore')
#returns movie name dictionary
    return movieNames
#makepairs is a funtion used to group data here as per user and ratings
def makePairs((user, ratings)):
#Extract movie rating pair- 1
    (movie1, rating1) = ratings[0]
#EExtract movie rating pair- 2
    (movie2, rating2) = ratings[1]
#return movies and rating as individual pairs
    return ((movie1, movie2), (rating1, rating2))

#remove duplicate
def filterDuplicates( (userID, ratings) ):
    (movie1, rating1) = ratings[0]
    (movie2, rating2) = ratings[1]
    return movie1 < movie2

#using cosinesimililarity to check the similarities between the rating pairs
def computeCosineSimilarity(ratingPairs):
#to keep a track of how many pairs we are computing
    numPairs = 0
    #x: first number in a pair;
    #y; second number in a pair;
    #xy= the product
    sum_xx = sum_yy = sum_xy = 0
    for ratingX, ratingY in ratingPairs:
#square of x,y and product of x&y as per the cosinesimililarity
        sum_xx += ratingX * ratingX
        sum_yy += ratingY * ratingY
        sum_xy += ratingX * ratingY
        numPairs += 1
#defining the formula
    numerator = sum_xy
    denominator = sqrt(sum_xx) * sqrt(sum_yy)

    score = 0
    if (denominator):
        score = (numerator / (float(denominator)))

    return (score, numPairs)

# now is when we work with spark and create RDD
conf = SparkConf().setMaster("local[*]").setAppName("MovieSimilarities") #creating the spark app name as "MovieSimilarities"
sc = SparkContext(conf = conf) #creating spark variable

print "\nLoading movie names..." #JUST TO INFORM
nameDict = loadMovieNames() #creating dictionary- calling loadmovienames

data = sc.textFile("file:///home/cloudera/moviedata/datafile2.txt") #SOURCE OF DATA: reading from textfile

#Converting a line of the data in the RDD to a list containing rating information
ratings = data.map(lambda l: l.split()).map(lambda l: (int(l[0]), (int(l[1]), float(l[2]))))

#Result: for each user, it gives all pairs of movies they've rated.
joinedRatings = ratings.join(ratings) #self-join

#Removes duplicate movie pairs and self-pairs (like (Movie1, Movie1)).
uniqueJoinedRatings = joinedRatings.filter(filterDuplicates)

#Converts joined data into actual movie pairs with their ratings.
moviePairs = uniqueJoinedRatings.map(makePairs)

#Groups all rating pairs for each unique movie pair.
moviePairRatings = moviePairs.groupByKey()

#Example Pair:((22, 35), (similarityScore, numberOfRatingsInCommon))
moviePairSimilarities = moviePairRatings.mapValues(computeCosineSimilarity).cache()

if (len(sys.argv) > 1): #condition for running next block

#setting the threshold
    scoreThreshold = 0.10
    coOccurenceThreshold = 2

#extracting movie ID
    movieID = int(sys.argv[1])

# Filtering movie pairs based on the set threshold
    filteredResults = moviePairSimilarities.filter(lambda((pair,sim)): \
        (pair[0] == movieID or pair[1] == movieID) \
        and sim[0] > scoreThreshold and sim[1] > coOccurenceThreshold)

#Extracting top 10 pairs based on the similarity scores
    results = filteredResults.map(lambda((pair,sim)): (sim, pair)).sortByKey(ascending = False).take(10)

#printing the final output as movie and similar movie set
    print "Top 10 similar movies for " + nameDict[movieID] #this is the original movie
    for result in results: #looping in all the top 10 similar movies against the original movie
        (sim, pair) = result
        similarMovieID = pair[0] #taking the first movie from the pair
        if (similarMovieID == movieID): #checking if the movie is same as original movie. If yes, then replace the similar movie ID from the pair.
            similarMovieID = pair[1]
        print nameDict[similarMovieID] + "\tscore: " + str(sim[0]) + "\tstrength: " + str(sim[1]) #print the similarity score and the similar movie ID
```
